```python
df1=df[['album_id','song_name']]
# df1=df1.drop_duplicates(keep='first')
df1
```

```python
data_dict=df1.to_dict('records')
print(len(data_dict))
print(data_dict[0])
indices=dict()

for doc in data_dict[:]:
#     print(doc)
   
        if doc['album_id'] not in indices:
            indices[doc['album_id']]=[]
        indices[doc['album_id']].append(str(doc['song_name']))
# indices

list_key_values=[[k,v] for k ,v in indices.items()]

df_if=pd.DataFrame.from_dict(list_key_values)
df_if.columns=['album_id','song_name']
df_if
```

```python
df_if['album_song']=[len(l) for l in df_if.song_name.values]
df_if.sort_values('album_song')
```
